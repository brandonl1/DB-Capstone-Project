Query with table JOIN

For the third and final task, Little Lemon need you to return specific details from your database. They require the full name and contact details for every customer that has placed an order greater than $60 for a promotional campaign. 

You can use the following steps to implement this functionality in your database directory:


Step One: Identify which tables are required. To complete the query, you first need to identify which table has the required data. 

The bill paid can be found in Orders as TotalCost and the customer contact information can be found in the Customers table. 

When selecting attributes from a specific table, specify the table name, followed by a dot and the target attribute as below (Hint: select the column attributes that you will need). 

```python
# SQL query stored in a string, selecting data from 'customers' and 'order' tables
Promotion = """
SELECT 
    customers.customer_first_name,  # Selects the first name of the customer
    customers.customer_last_name,   # Selects the last name of the customer
    customers.country_code,         # Selects the country code of the customer
    `order`.cost                    # Selects the cost from the 'order' table
FROM 
    customers                       # Data is selected from the 'customers' table
LEFT JOIN 
    `order`                         # Left join with the 'order' table
ON 
    customers.customer_id = `order`.customer_id  # Join condition based on customer ID
WHERE 
    `order`.cost > 60;              # Filter: Only include orders with cost greater than 60
"""

# Executing the SQL query using a cursor
cursor.execute(Promotion)

# Fetching all the results from the executed query
results = cursor.fetchall()

# Fetching the column names from the cursor
columns = cursor.column_names

# Printing the column names
print(columns)

# Printing each row from the results
for result in results:
    print(result)
```
